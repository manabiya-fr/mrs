```python
import numpy as np
from mrs.frames import RotationMatrix
from mrs.figs.rot_sequences import show_seq_rot, ax_config
from mrs.figs.utils import verify_bbox
```

```python
I = RotationMatrix()
rx = RotationMatrix.x(np.pi/2)
ry = RotationMatrix.y(np.pi/2)
rz = RotationMatrix.z(np.pi/2)
```

```python
ax = ax_config(show_seq_rot([I, (rx, 1.4), (ry@rx, 2.8), (rz@ry@rx, 4.1)]), elev=15, azim=-60)
verify_bbox(ax.figure, 1.15, 2.55, 5.5, 3.3, filename="/tmp/fig_verify_bbox.png")
```

```python
ax_config(show_seq_rot([I, (rz, 2.4), (rz@ry, 3.7), (rz@ry@rx, 4.1)]))
verify_bbox(ax.figure, 1.15, 2.55, 5.5, 3.3, filename="/tmp/fig_verify_bbox.png")
```
